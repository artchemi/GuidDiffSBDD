# 2. Свойства

```python
from rdkit import Chem
from rdkit.Chem import QED, Draw
from rdkit.Chem.Draw import MolsToGridImage
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import RDConfig

from PIL import Image
import io

sys.path.append(os.path.join(RDConfig.RDContribDir, 'SA_Score'))
from sascorer import sascorer
```

```python
ligands_gen = Path("test_samples/3rfm_gen.sdf")
```

```python
suppl = Chem.SDMolSupplier(str(ligands_gen)) 
ligands = [mol for mol in suppl if mol is not None]

smiles_list = [Chem.MolToSmiles(mol) for mol in ligands]
ligands_smiles = [Chem.MolFromSmiles(ligand_smi) for ligand_smi in smiles_list]

qed_values = [QED.qed(mol) for mol in ligands_smiles]
sas_values = [sascorer.calculateScore(mol) for mol in ligands]
labels = [f"QED: {qed:.4f}\nSAS: {sas:.4f}" for qed, sas in zip(qed_values, sas_values)]

img = MolsToGridImage(
    ligands_smiles, 
    molsPerRow=4, 
    subImgSize=(300,300), 
    legends=labels,
    )

display(img)
```

```python
import pymol2
from pathlib import Path
from PIL import Image
import imageio
from rdkit import Chem

# pdbfile = "protein.pdb"
# ligands_sdf = "ligands.sdf"
out_dir = Path("frames_pymol")
out_dir.mkdir(exist_ok=True)
gif_file = "ligands_animation.gif"

suppl = Chem.SDMolSupplier(str(ligands_gen))
ligands = [mol for mol in suppl if mol is not None]

with pymol2.PyMOL() as pymol:
    cmd = pymol.cmd
    cmd.set("ray_trace_frames", 1)
    cmd.set("ray_trace_mode", 1)
    cmd.bg_color("white")

    # === БЕЛОК ===
    cmd.load(pdbfile, "protein")
    cmd.hide("everything", "protein")
    cmd.show("cartoon", "protein")
    cmd.color("lime", "protein")

    # Выставляем ориентацию и сохраняем её
    cmd.zoom("protein")
    cmd.rotate("y", 160, "all")
    base_view = cmd.get_view()

    frame_files = []

    for i, ligand in enumerate(ligands):
        ligand_file = out_dir / f"ligand_{i:02d}.pdb"
        Chem.MolToPDBFile(ligand, str(ligand_file))

        cmd.load(str(ligand_file), "ligand")
        cmd.hide("everything", "ligand")
        cmd.show("sticks", "ligand")
        cmd.color("orange", "ligand")

        # Применяем сохранённый вид (чтобы белок не вращался)
        cmd.set_view(base_view)

        png_path = out_dir / f"frame_{i:02d}.png"
        cmd.png(str(png_path), width=800, height=600, dpi=300, ray=True)
        frame_files.append(png_path)
        print(f"Saved {png_path}")

        cmd.delete("ligand")

# === Собираем GIF ===
frames = [Image.open(f) for f in frame_files]
imageio.mimsave(gif_file, frames, fps=2)
print(f"✅ GIF saved as {gif_file}")
```
